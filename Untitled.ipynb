```python
import pandas as pd
from rl_agent import propose_strategy

# Charger les donn√©es connues
data = pd.read_csv("data/failures_data.csv")

# Transformer le solvant en identifiant num√©rique
solvent_map = {
    "water": 0,
    "ethanol": 1,
    "acetone": 2,
    "methanol": 3,
    "dichloromethane": 4
}

# Interface utilisateur via input()
print("üî¨ Enter your experiment parameters:")

temperature = int(input("‚Üí Temperature (¬∞C) : "))
solvent = input("‚Üí Solvent (water, ethanol, acetone, methanol, dichloromethane) : ").lower()
pH = float(input("‚Üí pH : "))
reaction_time = int(input("‚Üí Reaction time (min) : "))

# Formater la condition entr√©e
user_state = (
    temperature // 5,
    solvent_map.get(solvent, -1),
    int(pH),
    reaction_time // 5
)

# V√©rifier si cette condition existe dans les donn√©es connues
found = False
for i in range(len(data)):
    existing_state = (
        int(data["temperature_C"][i]) // 5,
        solvent_map[data["solvent"][i]],
        int(data["pH"][i]),
        int(data["reaction_time_min"][i]) // 5
    )
    if user_state == existing_state:
        found = True
        result = data["result"][i]
        status = "‚ùå FAILED" if result == 0 else "‚úÖ SUCCESS"
        print(f"\n‚ö†Ô∏è Match found in dataset ‚Üí Result: {status}")
        break

if not found:
    print("\n‚ÑπÔ∏è This experiment was not found in the dataset.")

# Proposer une alternative
print("\nü§ñ AI suggestion (safer conditions):")
suggestion = propose_strategy()
for key, value in suggestion.items():
    print(f"‚Üí {key}: {value}")
```

```python
# üîÑ Sauvegarder l'exp√©rience dans failures_data.csv
save = input("\nüíæ Do you want to save this experiment to the dataset? (yes/no): ").lower()

if save == "yes":
    new_entry = {
        "temperature_C": temperature,
        "solvent": solvent,
        "pH": pH,
        "reaction_time_min": reaction_time,
        "result": -1  # Valeur neutre, √† corriger manuellement plus tard
    }

    # Charger l'ancien fichier
    df = pd.read_csv("data/failures_data.csv")
    df = pd.concat([df, pd.DataFrame([new_entry])], ignore_index=True)

    # R√©√©crire le fichier CSV
    df.to_csv("data/failures_data.csv", index=False)

    print("‚úÖ Your experiment has been saved to failures_data.csv")
else:
    print("‚ùé Experiment not saved.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les donn√©es
df = pd.read_csv("data/failures_data.csv")

# Liste des solvants uniques
unique_solvents = df["solvent"].unique()

for solvent in unique_solvents:
    fig, ax = plt.subplots(figsize=(6, 4))
    subset = df[df["solvent"] == solvent]

    # √âchecs
    failures = subset[subset["result"] == 0]
    ax.scatter(failures["temperature_C"], failures["pH"], color='red', label='Failures')

    # R√©ussites
    successes = subset[subset["result"] == 1]
    ax.scatter(successes["temperature_C"], successes["pH"], color='green', label='Successes')

    # Mise en forme
    ax.set_xlabel("Temperature (¬∞C)")
    ax.set_ylabel("pH")
    ax.set_title(f"Results for Solvent: {solvent}")
    ax.legend()
    ax.grid(True)

    # Sauvegarder la figure
    fig.savefig(f"results_{solvent}.png", dpi=300, bbox_inches="tight")

    plt.show()
```
